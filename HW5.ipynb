```python
import jieba
import stanza
from collections import Counter, defaultdict
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# 文字雲與圖表字體路徑
font_path = "SourceHanSerifTC-Regular.otf"  # 確保這個字體檔案存在

# 設定 matplotlib 使用中文字體
my_font = fm.FontProperties(fname=font_path)
plt.rcParams['font.family'] = my_font.get_name()

# 初始化 stanza 中文模型
stanza.download('zh')  # 僅第一次執行需要
nlp = stanza.Pipeline('zh')

# 分析文本
text = """輝達（NVIDIA）執行長黃仁勳5月26日抵台，準備參加台北國際電腦展（COMPUTEX 2025）。他在台北101前現身，引發群眾圍觀與媒體爭相報導。黃仁勳此行除了演講，也預計拜訪台積電、廣達、華碩等科技公司，進一步洽談AI晶片合作事宜。台灣各界高度關注這位AI教父的到來，現場湧入大量科技迷與工程師。據悉，他將發表與生成式AI相關的重要發表，讓全球目光聚焦台灣，再次凸顯台灣在科技供應鏈中的關鍵角色。"""

# 斷詞處理
tokens = list(jieba.cut(text))
print("🔹 斷詞結果：")
print(tokens)

# NER 處理
doc = nlp(text)
print("\n🔹 命名實體識別結果：")
for ent in doc.ents:
    print(f"{ent.text} ({ent.type})")

# 高頻詞統計
stopwords = set(["，", "。", "的", "也", "他", "將", "與", "在", "前", "這位", "此行", "、", "。"])
filtered_tokens = [w for w in tokens if w not in stopwords and len(w) > 1]
word_freq = Counter(filtered_tokens)
top_words = word_freq.most_common(10)

# 繪製詞頻圖
words, freqs = zip(*top_words)
plt.figure(figsize=(10, 4))
plt.bar(words, freqs, color='skyblue')
plt.title("高頻詞統計", fontproperties=my_font)
plt.xlabel("詞語", fontproperties=my_font)
plt.ylabel("次數", fontproperties=my_font)
plt.xticks(rotation=45, fontproperties=my_font)
plt.tight_layout()
plt.show()

# 命名實體類型分佈圖
entity_counts = defaultdict(int)
for ent in doc.ents:
    entity_counts[ent.type] += 1

plt.figure(figsize=(6, 4))
plt.bar(entity_counts.keys(), entity_counts.values(), color='orange')
plt.title("命名實體類型分佈", fontproperties=my_font)
plt.xlabel("類型", fontproperties=my_font)
plt.ylabel("數量", fontproperties=my_font)
plt.xticks(fontproperties=my_font)
plt.tight_layout()
plt.show()
```
