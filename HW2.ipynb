```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

url = "https://raw.githubusercontent.com/datasets/global-temp/master/data/annual.csv"
df = pd.read_csv(url)

print(df['Source'].value_counts())

df_gcag = df[df['Source'] == 'gcag']


```

```python
df
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x='Year', y='Mean', data=df_gcag, color='tomato')
plt.title('全球年平均氣溫變化趨勢（GCAG）')
plt.xlabel('年份')
plt.ylabel('氣溫變化（°C）')
plt.grid(True)
plt.show()





```

```python
df_gcag['Decade'] = (df_gcag['Year'] // 10) * 10
decade_avg = df_gcag.groupby('Decade')['Mean'].mean().reset_index()

plt.figure(figsize=(10,6))
sns.barplot(x='Decade', y='Mean', data=decade_avg, palette='coolwarm')
plt.title('每 10 年全球平均氣溫變化（GCAG）')
plt.xlabel('年代')
plt.ylabel('平均氣溫變化（°C）')
plt.show()



```

```python
plt.figure(figsize=(12,6))

sns.lineplot(data=df[df['Source'] == 'gcag'], x='Year', y='Mean', label='GCAG', color='orange')
sns.lineplot(data=df[df['Source'] == 'GISTEMP'], x='Year', y='Mean', label='GISTEMP', color='blue')

plt.title('全球氣溫變化比較（GCAG vs GISTEMP）')
plt.xlabel('年份')
plt.ylabel('氣溫變化（°C）')
plt.legend()
plt.grid(True)
plt.show()


```
