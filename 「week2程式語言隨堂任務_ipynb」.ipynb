```python
food_calories = {
    "grains": {
        "Rice": 200,
        "Noodles": 220
    },
    "proteins": {
        "Egg": 70,
        "Tofu": 80
    }
}
```

```python
food_calories['proteins'].keys()
```

```python
import pandas as pd

df = pd.DataFrame(list(food_calories.keys()), columns=['Item'])
```

```python
df
```

```python
# 轉換為長格式
data_all = []
for category, items in food_calories.items():
    for food, calories in items.items():
        data_all.append([category, food, calories])

# 建立 DataFrame
df_all = pd.DataFrame(data_all, columns=['Category', 'Food', 'Calories'])
```

```python
df_all
```

```python
import pandas as pd
# read data and put it in a dataframe
# 在 google 工作表載入 gsheets
gsheets = gc.open_by_url('https://docs.google.com/spreadsheets/d/1M_VuXbmo3WRia7f0gY267j6ts53MYvIqAmObRbT10QA/edit?usp=sharing')
```

```python
from gspread_dataframe import set_with_dataframe

# 選擇要更新的工作表（選擇第一個工作表）
worksheet = gsheets.get_worksheet(0)  # 如果是其他工作表，可以用 get_worksheet(index)

# 將 DataFrame 寫入 Google Sheets
set_with_dataframe(worksheet, df_all)
```

```python
worksheet_money = gsheets.get_worksheet(1)
```

```python
worksheet_money
```

```python
# 取得 Google Sheets 內容
data = worksheet_money.get_all_values()  # 取得所有儲存格資料，回傳為 2D 陣列

# 將資料轉換為 DataFrame
df_money = pd.DataFrame(data)

# 設定第一行為標題（欄位名稱）
df_money.columns = df_money.iloc[0]  # 第一行設為欄位名稱
df_money = df_money[1:]  # 刪除第一行，保留數據部分

# 重設索引
df_money.reset_index(drop=True, inplace=True)
```

```python
df_money
```

```python
df_part = df_money[df_money['日期'] == '2025/02/26']
```
